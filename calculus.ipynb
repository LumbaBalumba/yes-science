```python
import sympy
```

```python
t, D, F, V_d, k_a, k_el, tau, alpha, X = sympy.symbols(r't D F V_d k_a k_el \tau \alpha X')
```

```python
absorption_model = F * D * k_a / V_d / \
    (k_a - k_el) * (sympy.exp(-k_el * t) - sympy.exp(-k_a * t))
absorption_model
```

```python
C = absorption_model.subs(t, tau)
elimination_model = (C * sympy.exp(-k_el * (t - tau))).simplify()
elimination_model
```

```python
model = sympy.Piecewise((absorption_model, t < tau), (elimination_model, t >= tau))
model
```

```python
loss = sympy.Piecewise(((X -  absorption_model) ** 2, t < tau), (alpha * (X - elimination_model)**2, t >= tau)).simplify()
loss
```

```python
f = sympy.Matrix([loss])
f
```

```python
J = f.jacobian([D, F, V_d, k_a, k_el])
J
```

```python
sympy.preview(J, viewer='file', filename='J.png')
```

```python
H = f.jacobian([D, F, V_d, k_a, k_el]).jacobian([D, F, V_d, k_a, k_el])
H
```

```python
sympy.preview(H, viewer='file', filename='H.png')
```
