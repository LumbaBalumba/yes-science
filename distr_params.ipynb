```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import lognorm
```

```python
samples = np.load('samples.npy')
```

```python
ta = []
for sample in samples:
    x, y = sample[0], sample[1]
    idx = np.argmax(y)
    ta.append(x[idx])
```

```python
plt.hist(ta, bins=25)
```

```python
shape, loc, scale = shape, loc, scale = lognorm.fit(ta)
```

```python
ta_values = np.linspace(np.min(ta), np.max(ta), 1000)
pdf_values = lognorm(shape, loc, scale).pdf(ta_values)
```

```python
plt.hist(ta, bins=25, density=True)
plt.plot(ta_values, pdf_values)
```
